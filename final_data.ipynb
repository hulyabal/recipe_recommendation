```python
# rating'e göre kategori oluştur
def categorize_rating(rating):
    if rating >= 4:
        return 'tasty'
    elif rating == 3:
        return 'so so'
    else:
        return 'bad'

# Yeni kategori sütununu oluştur
reviews['rating_category'] = reviews['rating'].apply(categorize_rating)

# Kategori sayısını göster
print(reviews['rating_category'].value_counts())
```

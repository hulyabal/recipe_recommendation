```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# TF-IDF vektörleme
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(df['ingredients'])

# Kosinüs benzerliği hesaplama
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Öneri fonksiyonu
def get_recommendations(recipe_index, cosine_similarities=cosine_sim):
    similar_recipes = list(enumerate(cosine_similarities[recipe_index]))
    # Benzerlik skorlarına göre sıralama
    similar_recipes = sorted(similar_recipes, key=lambda x: x[1], reverse=True)
    # İlk 5 benzer tarifi alma (çünkü kendisi hariç en yüksek benzerlik kendi tarifi olacaktır)
    similar_recipes = similar_recipes[1:6]
    # İlgili tarif indekslerini alma
    recipe_indices = [index[0] for index in similar_recipes]
    # İlgili tarifleri DataFrame'den alma
    recommended_recipes = df.iloc[recipe_indices]
    return recommended_recipes

# Örnek kullanım: 0. indeksteki bir tarife önerileri alalım
recommendations = get_recommendations(0)
print(recommendations[['name', 'ingredients']])
```

```python
# Örneğin, 10. indeksteki bir tarife önerileri alalım
recommendations_for_recipe_10 = get_recommendations(10)
print(recommendations_for_recipe_10[['name', 'ingredients']])
```

```python
# Örneğin, 10. indeksteki bir tarife önerileri alalım
recommendations_for_recipe_10 = get_recommendations(10)
print(recommendations_for_recipe_10[['name', 'ingredients']].reset_index(drop=True))
```

```python
# Örneğin, 10. indeksteki bir tarife önerileri alalım
recommendations_for_recipe_10 = get_recommendations(10)
result = recommendations_for_recipe_10[['name', 'ingredients']].reset_index(drop=True)

# Düz metin olarak yazdırma
for index, row in result.iterrows():
    print(f"{index + 1}. {row['name']}\n   {row['ingredients']}\n")
```
